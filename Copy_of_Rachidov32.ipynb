```python
!pip install transformers gradio accelerate

from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import gradio as gr

model_id = "sambanovasystems/SambaLingo-Arabic-Base"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto", torch_dtype="auto")

generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

def respond(prompt):
    out = generator(prompt, max_new_tokens=200, do_sample=True, temperature=0.7)
    return out[0]["generated_text"]

gr.Interface(
    fn=respond,
    inputs="text",
    outputs="text",
    title="SambaLingo Arabic 7B – بدون رقابة",
    description="أدخل النص بالعربي أو الإنجليزي وسيتم الرد بدون فلترة."
).launch(share=True)
```
